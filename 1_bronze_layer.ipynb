```python
#Importing libraries for connecting to API
import requests
import json
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('police_incidents').getOrCreate()
```

```python
#Extracting data that is available in real time from the last 500 incidents
url = 'https://polisen.se/api/events'

response = requests.get(url)

data = response.json()

dfspark = spark.createDataFrame(data)
```

```python
dfspark.printSchema()
```

```python
dfspark.display()
```

```python
#Appending the data to the bronze delta table
dfspark.write.format('delta').mode('append').saveAsTable('thesis_police.bronze.incidents')
```

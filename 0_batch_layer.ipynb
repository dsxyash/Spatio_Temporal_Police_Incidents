```python
#Importing libraries for connecting to API
import requests
import json
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName('police_incidents').getOrCreate()
```

```python
#Extracting data from the last month that is available to extract from, which is november
url = 'https://polisen.se/api/events?DateTime=2025-11'

response = requests.get(url)

data = response.json()

dfspark = spark.createDataFrame(data)
```

```python
dfspark.printSchema()
```

### Saving november 2025

```python
#Saving the data to the volume
dfspark.coalesce(1).write.mode('overwrite').parquet('/Volumes/thesis_police/bronze/raw_data/incidents_2025_11.parquet')
```

```python
#Appending the raw data to the bronze delta table
dfspark.write.format('delta').mode('append').saveAsTable('thesis_police.bronze.incidents')
```
